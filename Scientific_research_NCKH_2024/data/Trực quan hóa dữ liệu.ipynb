```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Xử lý dữ liệu: nhóm sản phẩm theo danh mục (category_id) và đếm số lượng
category_counts = df['category'].value_counts()

# Trực quan hóa: Biểu đồ cột cho số lượng sản phẩm theo danh mục
plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index, y=category_counts.values, palette="viridis")
plt.title("Số lượng sản phẩm theo danh mục", fontsize=14)
plt.xlabel("Danh mục (Category)", fontsize=12)
plt.ylabel("Số lượng sản phẩm", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Xử lý dữ liệu: Đảm bảo cột 'final_price' có định dạng số
df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')

# Lọc dữ liệu: Loại bỏ các giá trị NaN và thương hiệu không xác định (UNKNOWN)
filtered_data = df.dropna(subset=['final_price', 'brand'])
filtered_data = filtered_data[filtered_data['brand'].str.upper() != 'UNKNOWN']

# Trực quan hóa: Phân phối giá trung bình theo thương hiệu (biểu đồ hộp)
plt.figure(figsize=(12, 8))
top_brands = filtered_data['brand'].value_counts().head(10).index  # Chỉ lấy top 10 thương hiệu
sns.boxplot(data=filtered_data[filtered_data['brand'].isin(top_brands)], 
            x='brand', y='final_price', palette="Set3")
plt.title("Phân phối giá theo thương hiệu (Top 10)", fontsize=14)
plt.xlabel("Thương hiệu", fontsize=12)
plt.ylabel("Giá (final_price)", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Xử lý dữ liệu: Đảm bảo cột 'final_price' có định dạng số
df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')

# Lọc dữ liệu: Loại bỏ các giá trị NaN và thương hiệu không xác định (UNKNOWN)
filtered_data = df.dropna(subset=['final_price', 'brand'])
filtered_data = filtered_data[filtered_data['brand'].str.upper() != 'UNKNOWN']

# Loại bỏ các giá trị cực trị: Áp dụng quy tắc IQR (Interquartile Range)
Q1 = filtered_data['final_price'].quantile(0.25)
Q3 = filtered_data['final_price'].quantile(0.75)
IQR = Q3 - Q1

# Lọc các giá trị ngoài phạm vi (Q1 - 1.5 * IQR, Q3 + 1.5 * IQR)
filtered_data_no_outliers = filtered_data[(filtered_data['final_price'] >= (Q1 - 1.5 * IQR)) & 
                                           (filtered_data['final_price'] <= (Q3 + 1.5 * IQR))]

# Trực quan hóa lại: Biểu đồ hộp cho các thương hiệu sau khi loại bỏ cực trị
plt.figure(figsize=(12, 8))
top_brands = filtered_data_no_outliers['brand'].value_counts().head(10).index  # Chỉ lấy top 10 thương hiệu
sns.boxplot(data=filtered_data_no_outliers[filtered_data_no_outliers['brand'].isin(top_brands)], 
            x='brand', y='final_price', palette="Set3")
plt.title("Phân phối giá theo thương hiệu (Top 10, loại bỏ cực trị)", fontsize=14)
plt.xlabel("Thương hiệu", fontsize=12)
plt.ylabel("Giá (final_price)", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
